```python
import requests
import bs4
import pandas as pd
```

```python
request1 = requests.get("https://www.flipkart.com/moto-g34-5g/p/itm38a4f74f43133?pid=MOBGUFK49KDBZUMC&ocmpid=BrandAd_Motorola_g34_Google_Search_India-Launch-ProdSpec-Paid?force_app=1&gad_source=1&gclid=Cj0KCQjw_-GxBhC1ARIsADGgDjsezrvnQf6Sxb2tgNeLcz0RKXG1KiGtIz8tM0s1A0dGuOBGctJ_0YMaApm9EALw_wcB")
request1
```

```python
request1.content
```

```python
Soup = bs4.BeautifulSoup(request1.text)
Soup
```

Fetching Reviews and comments

```python
reviews = Soup.findAll('div' ,{'class' : 'ZmyHeo'});
for review in reviews:
    print(review.get_text() + "\n\n")
```

Fetching overall Rating

```python
rating = Soup.find('div',{'class' : 'ipqd2A'}).get_text();
print(rating)
```

```python
individual_ratings = Soup.findAll('div',{'class' : 'XQDdHH Ga3i8K'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text( )+ "\n")
```

```python
tags = Soup.find('span',{'class' : '_6EBuvT OHdLLG'}).get_text();
tags
```

Fetching Customer Names

```python
customer_name = Soup.findAll('p',{'class':'_2NsDsF AwS1CA'});
for cust_name in customer_name:
    print(cust_name.text.strip() + "\n")
```
